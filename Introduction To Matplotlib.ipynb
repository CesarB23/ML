##### Matplotlib is built on Numpy arrays and Python, is integrated directly with pandas

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

#Implicit Pyplot Style
x = np.linspace(-3*(np.pi),3*(np.pi),100)
y = np.linspace(0,np.pi,100)
x2= np.cos(x)
x = np.sin(x)

plt.title("Sin (x) and Cos (x)")
plt.xlabel("X")
plt.ylabel("Y",loc="center")
fig = plt.plot(y,x,y,x2)
```

```python
#OO-Style
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
heart_df = pd.read_csv("heart-disease.csv")
heart_df
```

```python
heart_df.info()
```

```python
heart_df["age"].plot.hist(bins=10)
```

```python
heart_df.head()
```

```python
fix,((ax1,ax2),(ax3,ax4)) = plt.subplots(ncols=2,
                                        nrows=2,
                                        figsize=(10,8))
ax1.hist(heart_df["age"],bins=5)
ax2.hist(heart_df["trestbps"],bins=5)
ax3.hist(heart_df["thalach"],bins=5)
ax4.hist(heart_df["oldpeak"],bins=5);
```

```python
over_50 = heart_df[heart_df["age"] >50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
ax.set_xlim([50,80]);
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
fig,(ax1,ax2) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True)

scatter = ax1.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
scatter_2 = ax2.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")

ax1.set(title="Heart Disease And Cholesterol Levels",
        ylabel="Cholesterol");
ax1.set_xlim([50,80])
ax2.set_ylim([60,200])
ax2.set(title="Heart Disease and Max Heart Rate",
       ylabel="Max Heart Rate",
        xlabel="Age")

ax1.legend(*scatter.legend_elements(),title="Target")
ax2.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_50["chol"].mean(),ls='-.');
ax2.axhline(over_50["thalach"].mean(),ls='--');

fig.suptitle("Heart Disease Analysis",fontsize=20,fontweight="bold");
```
