```python
#Comentarios de Ejemplo

import numpy as np
import matplotlib.pyplot as plt
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                             ncols=2,
                             figsize=(10,5))

#Graficas de multuples funciones matematicas en un cuadricula de 2*2
x = np.linspace(-3*(np.pi),3*(np.pi),100)
y = np.linspace(0,np.pi,100)
ax1.plot(y,np.sin(x))
ax2.plot(y,np.cos(x))
ax3.plot(y,np.cos(x)*np.sin(x))
ax4.plot(y,np.exp(x));
```
